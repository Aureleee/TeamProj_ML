```python
# importation of depedencies

import pandas as pd 
import numpy as np
```

## Loading data

```python
Kma_data= pd.read_csv("official_weather_data/weather_train.csv")
Kma_data.rename(columns={"Unnamed: 0": "id"}, inplace=True)
```

```python
Kma_data
```

```python
Kma_data.isna().sum()
# We can see that there's a missing data nan
```

### Normalisation

```python
#Non-features columns ; NF_columns  -> ['id','day',"rainfall"]
features=Kma_data.drop(columns=['id','rainfall'])
#Target ; ["rainfall"]
targets=Kma_data["rainfall"]

x=features.values
means=x.mean(axis=0) #mean for each columns
std=x.std(axis=0)
x_normalize= (x-means)/std
```

```python
data_train_n=Kma_data.copy()
data_train_n[features.columns]=x_normalize
data_train_n.to_csv("official_weather_data/kma_train.csv",index=False)
data_train_n
```

#### test norm

```python
Kma_data_t= pd.read_csv("official_weather_data/weather_test.csv")
Kma_data_t.rename(columns={"Unnamed: 0": "id"}, inplace=True)
```

```python
Kma_data_t
```

```python
X=Kma_data_t.drop(columns=['id','rainfall'])
f=X.values
m,s=f.mean(axis=0),f.std(axis=0)
f_normalise=(X-m)/s
data_test_n=Kma_data_t.copy()
data_test_n[X.columns]=f_normalise
data_test_n.to_csv("official_weather_data/kma_test.csv",index=False)
data_test_n
```
